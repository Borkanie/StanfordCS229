# Binomial Distribution

We are interested in the number of times the event $A$ occurs in $n$ independent trials. $x$ is the number of times occurs in trials.
Noting the probability of the event $A$ to appear at a given trial as $p$ , $P(A)=p$ we note with $q$ the probability of this event not to trigger.
Thus $q=P(!A)=1-p$. If the trials are independent and the probaiblity of the vent does not change we can conlucle that: \
$  P({N=x}) =
    \left\{ \begin{array}{l}
     ({\begin{array}{l}
     n \\
     x \end{array}})*p^{x}*q^(n-x) \space | \space x \space \epsilon \space \overline{1,n} \\
     0 ,otherwise
    \end{array} \right.
     $


# Gaussian Distribution

The gaussian distribution, also known as a bell curve, can be defined using a probabilitry distribution function of parameters mean $\mu$ and variance $\sigma^{2}$ with the following expression:\
$
P(A) = \frac {e^\frac {(x-\mu)^2} {2*\sigma^2} } {\sqrt{2*\sigma*\pi}}
$

```python
import numpy as np
import matplotlib.pyplot as plot
import matplotlib
import math

class BooleanBinomialDistribution:
    def __init__(self,probailityOfEvent):
        self.p = probailityOfEvent
        
    def probability(self,numberOftrials,wantedResults):
        return math.comb(numberOftrials,wantedResults)*self.p**wantedResults*(1 - self.p)**(numberOftrials - wantedResults)
class GaussianDistribution:
    def __init__(self,mean,variance):
        self.mean = mean
        self.variance=variance
        
    def probability(self,x):
        return math.exp(-((x-self.mean)**2)/(2*self.variance))/(math.sqrt(self.variance*2*math.pi))
    

probability=0.5
n=25
eventGaus = GaussianDistribution(mean=n*probability,variance=n/2)
eventBinom = BooleanBinomialDistribution(probability)

resultGaus = []
resultBinom = []


for i  in np.arange(0,n):
    resultBinom.append(eventBinom.probability(n,i))
    resultGaus.append(eventGaus.probability(i))
plot.stem(resultBinom)
plot.stem(resultGaus,linefmt='C1-')
plot.xlabel('X=x')
plot.ylabel('Probability')
plot.legend({"Binomial","Gaussian"})
plot.show()
```
